```python
req = urllib.request.Request('https://www.anekdot.ru/tags/')
response = urllib.request.urlopen(req)
html_content = response.read().decode('utf-8')
```

```python
for line in html_content.splitlines():
    if '>Тема<' in line:
        topics_line = line
```

```python
links0 = re.findall(r'<tr><td><a href=".*?" title', topics_line)
```

```python
numbers0 = re.findall(r'>\d*?<', topics_line)
```

```python
numbers1 = []
for i in numbers0:
    if i != '><':
        numbers1.append(i[1:-1])
```

```python
len(numbers1)
```

```python
num_of_pages = list(map(lambda x: int(x) // 50 + 1, numbers1))
```

```python
links = []
for link in links0:
    try:
        newlink = 'https://www.anekdot.ru' + re.sub('"', '', re.search(r'".*?"', link).group(0))
        links.append(newlink)
    except:
        pass
```

обходим все страницы темы

```python
def link_generator(links, num_of_pages):
    all_links = []
    for i in range(len(links)):
        all_links.append(links[i])
        for j in range(2, num_of_pages[i]+1):
            all_links.append(links[i]+'/'+str(j))
    return all_links
```

```python
all_links = link_generator(links, num_of_pages)
len(all_links)
```

начинаем скачивать анекдоты

```python
def crawl_link(link):
    req = urllib.request.Request(link)
    response = urllib.request.urlopen(req)
    strings = response.read().decode('utf-8').splitlines()

    content = []
    for string in strings:
        if 'votingbox' in string:
            content.append(string)
    
    almost_anecs = []
    for line in content:
        almost_anecs.extend(line.split('votingbox'))

    anecs = []
    for anec in almost_anecs:
        try:
            text = re.search(r'''<div class="text">.*?</div>''', anec).group(0)
            text = re.sub(r'<.*?>', '', text)
            anecs.append(text)
        except:
            pass

    return anecs
```

```python
def anecs_to_file(filename, anecs):
    with open(filename, 'a', encoding='utf-8') as f:
        for text in anecs:
            f.write(text)
            f.write('\n sep@rator \n')
```

```python
for i in range(1400, 1848):
    anecs_to_file('sixth.txt', crawl_link(all_links[i]))
```
